# Tidy data

## df1: filter rows of which 'mainText' column is empty

```python
df_tidy1 = df_raw1['_source'].apply(pd.Series)
df_tidy1_new = df_tidy1[['court','date','judgement','mainText']]
df_tidy1_new['judgement'].replace('\s+', '', regex=True, inplace=True)
df_tidy1_new['mainText'].replace('\s+', '', regex=True, inplace=True)
df_tidy1_new
```

```python
for text in df_tidy1_new['mainText'][-1199:]:
    print(text)
```

```python
df_tidy1_new['mainText'].replace('', np.nan, inplace=True)
df_tidy1_new.dropna(subset=['mainText'], inplace=True)
df_tidy1_new
```

## df2: add "mainText", filter rows which are not 刑事判決/公然侮辱

```python
df_tidy2 = df_raw2['_source'].apply(pd.Series)
df_tidy2_new = df_tidy2[['court','date','judgement']]
df_tidy2_new['judgement'].replace('\s+', '', regex=True, inplace=True)
df_tidy2_new
```

```python
import re
df_tidy2_new['mainText'] = df_tidy2_new['judgement'].str.extract('(主文.*(?=事實及理由))')
df_tidy2_new['mainText'] = df_tidy2_new['mainText'].str.replace(r'主文', '')
df_tidy2_new['mainText'].replace('', np.nan, inplace=True)
df_tidy2_new.dropna(subset=['mainText'], inplace=True)
df_tidy2_new
```

```python
#鎖定累犯註記、判刑（拘役、有期徒刑）用詞
a1 = "[，,，](累犯)*(共*罪)*[，,，]*[各均有]*處*[,，]*(有期徒*刑徒*|拘役)"
a2 = "[，,，](累犯)*(共*罪)*[，,，]*[各均有]*[處科]*[,，]*罰金(新[台臺]幣)*"


#鎖定處刑用詞
b1 = "[一二三四五六七八九十壹貳參肆伍陸柒捌玖拾廿卅１２３４５６７８９０]*(年|個*月|日)"
b2 = "[一二三四五六七八九十壹貳參肆伍陸柒捌玖拾廿卅百佰千仟萬１２３４５６７８９０]*元*"


#找出剝奪自由刑之模式組合
findimprison = a1+b1


#找出罰金之模式組合
findf = a2+b2

#無罪的模式組合
c = "[，,，、，]*部*[分份]*[，,，、，]*均*無罪"
findn = c


#把非刑期、罰金的描述都刪掉（很暴力）
rep = "情形|罰金|科|處|新[台臺]幣|銀|元|不能安全|駕駛|動力|交通工具|而|吐氣|所含|酒精濃度|達|每公升.*毫克|以上|而|者|,|，|、.*|；.*|。.*|累犯|共.罪|有期徒*刑徒*|罪|拘役|過失|如易科.*|被訴.*|應執行.*|緩刑.*|各|均|部[份分].*|共|有|之"
```

```python
result_imprison_text = []
result_penalty_text = []

for i in swear_edit['mainText']:    
    if bool(re.search(findimprison, i)):
        r  = re.sub(rep,"",re.search(findimprison, i).group())
        result_imprison_text.append(r)
        result_penalty_text.append('')
        
    elif bool(re.search(findf,i)):
        r  = re.sub(rep,"",re.search(findf,i).group())
        result_imprison_text.append("")
        result_penalty_text.append(r)    
        
    elif bool(re.search(findn,i)):
        result_imprison_text.append("無罪")
        result_penalty_text.append("")        
    else:
        result_imprison_text.append("")
        result_penalty_text.append("")
print(len(result_imprison_text))
print(len(result_penalty_text))
```

```python
def nds(y):
    if bool(re.search('[壹貳參肆伍陸柒捌玖拾廿卅佰百仟千萬一二三四五六七八九十１２３４５６７８９０]', y)):
        if len(y) == 1:
            return text2num[y]
        elif len(y) == 2:
            if y[1]=='拾':
                return text2num[y[0]]*10
            elif y[0] in ['拾', '廿', '卅']:
                return text2num[y[0]] + text2num[y[1]]
            else:
                return text2num[y[0]]*10 + text2num[y[1]]
        elif len(y) == 3:
            return text2num[y[0]] * text2num[y[1]] + text2num[y[2]]

        else:
            return text2num[y]
    else:
        return 0
```

```python
def todays(x):
    if '年' in x or '月' in x or '日' in x:
        try:
            y = x.index("年")
        except:
            y = 0
        try:
            m = x.index("月")
        except:
            m = 0
        try:
            d = x.index("日")
        except:
            d = 0

        if y>0:
            year = x[0:y]

            if m>0:
                month = x[0+y+1:m]

                if d>0:
                    day = x[0+m+1:d]
                else:
                    day = ""
            else:
                month = ""

                if d>0:
                    day = x[0+y+1:d]
                else:
                    day = ""
        else:
            year = ""

            if m>0:
                month = x[0:m]

                if d>0:
                    day = x[0+m+1:d]

                else:
                    day =""
            else:
                month = ""
                day = x[0:d]

        res = nds(year)*365+nds(month)*30+nds(day)
        return res

    else:
        return 0
```
